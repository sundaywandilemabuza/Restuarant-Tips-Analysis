```python
## Visualization

import pandas as pd

url="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df=pd.read_csv(url)

import matplotlib.pyplot as plt

df.groupby("day")["tip"].mean().plot(kind="bar")
plt.xlabel("Day of the Week")
plt.ylabel("Average Tip")
plt.title("Average Tip per Day")
plt.show()
```

###Insights
- Tipping behavoir varies across the week, with certain days showing stronger average tipping pattern. This suggest that customer mindset and dining purpose may differ depending on the day.

```python
df.groupby("size")["tip"].mean().plot(kind="bar")
plt.xlabel("Group Size")
plt.ylabel("Average Tip")
plt.title("Average Tip per Group Size")
plt.show()
```

###Insights
- Mid-sized and larger groups tend to contribute significantly to hihger aveage tips. This indicates thet group dining dynamics play an important role in tipping behavoir.

```python
plt.scatter(df["total_bill"],df["tip"])
plt.title("Tip vs Total Bill")
plt.xlabel("Total bill")
plt.ylabel("Tip")
plt.show()
```

###Insights
- There is a clear positive relationship between total bill and tip amount. customes who spend more tend to leave higher tips, indicating that increasing customer spending can directly influence tipping revenue.
